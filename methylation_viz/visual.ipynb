```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the datasets
normal = pd.read_csv("meg3_methylation_norm.csv")
diabetic = pd.read_csv("meg3_methylation_diab.csv")
```

```python
print(normal.columns)
```

```python
# Plotting
plt.figure(figsize=(12,6))
# plt.plot(normal['position'], normal['percent_methylation'], label='Normal', color='blue')
# plt.fill_between(normal['position'], normal['percent_methylation'], alpha=0.2, color='blue')

# plt.plot(diabetic['position'], diabetic['percent_methylation'], label='Diabetic', color='red')
# plt.fill_between(diabetic['position'], diabetic['percent_methylation'], alpha=0.2, color='red')

plt.bar(normal['position'], normal['percent_methylation'], width=2, label='Normal', alpha=0.4, color='blue')
plt.bar(diabetic['position'], diabetic['percent_methylation'], width=2, label='Diabetic', alpha=0.2, color='red')

plt.grid(True, linestyle='--', alpha=0.3)

plt.xlabel("Genomic position within MEG3 locus")
plt.ylabel("Percent methylation")
plt.title("MEG3 promoter methylation: Normal vs Diabetic")
plt.legend()
# plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Create figure and axes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Plot for Normal
ax1.bar(normal['position'], normal['percent_methylation'], color='blue', alpha=0.7)
ax1.set_title("Normal Sample")
ax1.set_xlabel("CpG Position")
ax1.set_ylabel("Percent Methylation")
ax1.grid(axis='y', linestyle='--', alpha=0.3)
ax1.set_xticks(normal['position'])
ax1.set_xticklabels(normal['position'], rotation=45)

# Plot for Diabetic
ax2.bar(diabetic['position'], diabetic['percent_methylation'], color='red', alpha=0.7)
ax2.set_title("Diabetic Sample")
ax2.set_xlabel("CpG Position")
ax2.grid(axis='y', linestyle='--', alpha=0.3)
ax2.set_xticks(diabetic['position'])
ax2.set_xticklabels(diabetic['position'], rotation=45)

# Adjust layout
plt.suptitle("CpG Methylation at MEG3 Locus")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(normal['position'], normal['percent_methylation'], label='Normal', color='blue', marker='o')
plt.fill_between(normal['position'], normal['percent_methylation'], alpha=0.2, color='blue')

plt.plot(diabetic['position'], diabetic['percent_methylation'], label='Diabetic', color='red', marker='o')
plt.fill_between(diabetic['position'], diabetic['percent_methylation'], alpha=0.2, color='red')

plt.grid(True, linestyle='--', alpha=0.3)

plt.xlabel("Genomic position within MEG3 locus")
plt.ylabel("Percent methylation")
plt.title("MEG3 promoter methylation: Normal vs Diabetic")
plt.legend()
# plt.grid(True)
plt.tight_layout()
plt.show()
```
